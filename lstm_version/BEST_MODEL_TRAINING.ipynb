```python
def pitch_class_vector(notes):
    vec = np.zeros(12)
    for note in notes:
        vec[note % 12] += 1
    return vec

def interval_based_vector(notes):
    if len(notes) <= 1:
        return np.zeros(12)  # No intervals with 0 or 1 note
        
    notes = sorted(notes)
    intervals = []
    
    # Get all intervals between pairs of notes
    for i in range(len(notes)):
        for j in range(i+1, len(notes)):
            intervals.append((notes[j] - notes[i]) % 12)
    
    # Count interval occurrences
    interval_vector = np.zeros(12)
    for interval in intervals:
        interval_vector[interval] += 1
    
    # Normalize
    if np.sum(interval_vector) > 0:
        interval_vector = interval_vector / np.sum(interval_vector)
        
    return interval_vector

"""def best_chord_match(notes, chord_templates, verbose=False):
    input_set = set(notes)
    formatted_notes = pitch_class_vector(notes).reshape(-1, 1)
    print(f"Notes vector: {formatted_notes}")

    scores = [[0, 0] for x in range(24)]
    
    # Second pass: compute cosine similarity for candidates with any overlap
    for chord_label, chord_notes in chord_templates.items():
        template_set = set(chord_notes)
        if len(input_set & template_set) == 0:
            continue  # skip chords with no shared notes

        formatted_chord = pitch_class_vector(chord_notes).reshape(-1, 1)
        print(f"Chord vector: {formatted_chord}")
        curr_score = cosine_similarity(formatted_notes, formatted_chord)[0][0]
        overlap_size = 1/(5 - len(input_set & template_set))

        if verbose:
            print(f"Comparing with {chord_label}: cosine={curr_score:.3f}, overlap={overlap_size}")

        scores[chord_label][0] = curr_score
        scores[chord_label][1] = overlap_size

    combined_scores = [ cos * overlap for cos, overlap in scores ]
    
    return combined_scores"""

def best_chord_match(notes, chord_templates, verbose=False):
    if not notes:
        return np.zeros(24)
    
    # Create normalized pitch class vector
    input_vector = pitch_class_vector(notes)
    if np.sum(input_vector) > 0:
        input_vector = input_vector / np.sum(input_vector)
    
    # Also create interval vector for additional feature
    interval_vec = interval_based_vector(notes)
    
    scores = np.zeros(24)
    
    for chord_label, chord_notes in chord_templates.items():
        # Create chord template vector and normalize
        chord_vector = pitch_class_vector(chord_notes)
        if np.sum(chord_vector) > 0:
            chord_vector = chord_vector / np.sum(chord_vector)
        
        # Chord interval vector
        chord_interval_vec = interval_based_vector(chord_notes)
        
        # Compute pitch class similarity
        pitch_similarity = cosine_similarity(
            input_vector.reshape(1, -1), 
            chord_vector.reshape(1, -1)
        )[0][0]
        
        # Compute interval similarity
        interval_similarity = cosine_similarity(
            interval_vec.reshape(1, -1),
            chord_interval_vec.reshape(1, -1)
        )[0][0] if len(notes) > 1 else 0
        
        # Combine similarities (can adjust weights)
        combined_score = 0.7 * pitch_similarity + 0.3 * interval_similarity
        
        scores[chord_label] = combined_score
        
        if verbose and chord_label < 5:  # Limit verbose output to first few chords
            print(f"Chord {chord_label}: pitch_sim={pitch_similarity:.3f}, "
                  f"interval_sim={interval_similarity:.3f}, score={combined_score:.3f}")
    
    return scores
```

```python
# --- Setup ---
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Dense, Masking, Input, Embedding, Concatenate
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity

# --- Constants ---
NUM_MIDI_NOTES = 128
NUM_CHORDS = 24
CHORD_INPUT_DIM = NUM_CHORDS + 1  # +1 for 'None'
INPUT_DIM = NUM_MIDI_NOTES + CHORD_INPUT_DIM

# --- Utility Functions ---
def notes_to_vector(notes):
    vec = np.zeros(NUM_MIDI_NOTES)
    for n in notes:
        if 0 <= n < NUM_MIDI_NOTES:
            vec[n] = 1
    return vec

def build_input_sequence(sequences):
    note_seqs, chord_seqs, y_seqs = [], [], []
    for seq in sequences:
        note_seq, chord_seq, y_seq = [], [], []
        for notes, prev_chord, current_chord in seq:
            note_vec = notes_to_vector(notes)
            chord_idx = prev_chord if prev_chord is not None else NUM_CHORDS
            note_seq.append(note_vec)
            chord_seq.append(chord_idx)
            y_seq.append(current_chord)
        note_seqs.append(note_seq)
        chord_seqs.append(chord_seq)
        y_seqs.append(y_seq)

    note_padded = pad_sequences(note_seqs, padding='post', dtype='float32')
    chord_padded = pad_sequences(chord_seqs, padding='post', dtype='int32')
    y_padded = pad_sequences(y_seqs, padding='post', dtype='int32')
    return note_padded, chord_padded, y_padded
    
# --- Train/Test Split ---
train_data, test_data = train_test_split(all_data, test_size=0.2) # random_state=42

# --- Vectorize ---
note_train, chord_train, y_train = build_input_sequence(train_data)
note_test, chord_test, y_test = build_input_sequence(test_data)

# --- Model ---
note_input = Input(shape=(None, NUM_MIDI_NOTES), name='notes')
chord_input = Input(shape=(None,), dtype='int32', name='prev_chords')

# Embed previous chord (25 possible values → 16-dim vectors)
chord_embedding = Embedding(input_dim=CHORD_INPUT_DIM, output_dim=16)(chord_input)

# Combine note vector and chord embedding
combined_input = Concatenate()([note_input, chord_embedding])

x = Masking(mask_value=0.0)(combined_input)
x = LSTM(128, return_sequences=True)(x)
output = Dense(NUM_CHORDS, activation='softmax')(x)

model = Model(inputs=[note_input, chord_input], outputs=output)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# --- Train ---
model.fit(
    [note_train, chord_train],
    y_train,
    epochs=10,
    batch_size=8,
    validation_split=0.2
)
```

```python
# --- Inference Function ---
def predict_chords(model, note_sequence, power=1):
    note_inputs = []
    chord_inputs = []
    predictions = []
    confidences = []
    prev_chord = NUM_CHORDS  # 'None'

    for notes in note_sequence:
        note_vec = notes_to_vector(notes)
        note_inputs.append(note_vec)
        chord_inputs.append(prev_chord)

        X_notes = np.array(note_inputs)[np.newaxis, :, :]
        X_chords = np.array(chord_inputs)[np.newaxis, :]
        pred = model.predict([X_notes, X_chords], verbose=0)[0, -1]
        max_pred_conf = float(np.max(pred))
        deterministic = best_chord_match(notes, midi_chord_dict)
        
        cumulative_scores = [p * (max_pred_conf ** power) + det * (1 - max_pred_conf ** power) for p, det in zip(pred, deterministic)]
        prev_chord = np.argmax(cumulative_scores)
        confidence = float(np.max(cumulative_scores))

        predictions.append(prev_chord)
        confidences.append(confidence)

    return predictions, confidences

def model_only(model, note_sequence):
    note_inputs = []
    chord_inputs = []
    predictions = []
    confidences = []
    prev_chord = NUM_CHORDS  # 'None'

    for notes in note_sequence:
        note_vec = notes_to_vector(notes)
        note_inputs.append(note_vec)
        chord_inputs.append(prev_chord)

        X_notes = np.array(note_inputs)[np.newaxis, :, :]
        X_chords = np.array(chord_inputs)[np.newaxis, :]
        pred = model.predict([X_notes, X_chords], verbose=0)[0, -1]

        prev_chord = np.argmax(pred)
        confidence = float(np.max(pred))

        predictions.append(prev_chord)
        confidences.append(confidence)

    return predictions, confidences


def deterministic_only(model, note_sequence):
    predictions = []
    confidences = []
    prev_chord = NUM_CHORDS  # 'None'

    for notes in note_sequence:
        deterministic = best_chord_match(notes, midi_chord_dict)
        pred = np.argmax(deterministic)
        confidence = float(np.max(deterministic))

        predictions.append(pred)
        confidences.append(confidences)

    return predictions, confidences

def hamming_distance(list1, list2):
    if len(list1) != len(list2):
        raise ValueError("Lists must be of the same length")
    return sum(el1 != el2 for el1, el2 in zip(list1, list2))

for name, chord in chords.items():
    pred_chord, conf = predict_chords(model, [chord])
    print(f"{name}, {chord_encodings[pred_chord[0]]}, {conf}")

'''for name, chord in chords.items():
    pred_chord = best_chord_match(chord, midi_chord_dict)
    print(f"{name}, {chord_encodings[pred_chord]}")'''
```
